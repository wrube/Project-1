```python
import pandas as pd
import re
```

```python
disasters_df = pd.read_csv("./input_data/natural-disasters.csv")

# Specify keywords
keywords = ['country', 'year', 'earthquake']
regex_pattern = re.compile('|'.join(keywords), flags=re.IGNORECASE)

# Select columns containing the specified keywords
disasters_filtered_df = disasters_df.filter(regex=regex_pattern, axis=1)

disasters_filtered_df = disasters_filtered_df.loc[disasters_filtered_df["Year"] >= 2010 ]

# Display the selected columns
disasters_filtered_df.head(50)
```
